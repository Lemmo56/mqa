# Manipulando track_details

```python
# Lendo o dataset com os detalhes das m√∫sicas
df_track_details = pd.read_csv('datasets/source/track_details.csv')
```

```python
df_track_details
```

```python
# Conferindo se h√° colunas com valor nulo e o tipo dos dados de cada coluna
df_track_details.info()
```

```python
# Transformando o mode em num√©rico para ser poss√≠vel calcular a moda
df_track_details['mode_numeric'] = df_track_details['mode'].map({"minor": 0, "major":1})
```

```python
# Separando as vari√°veis que usaremos (as quantitativas usam a m√©dia enquanto as qualitativas usam a moda)
df_track_details_quantitative_variables = df_track_details[['track_id', 'duration', 'us_popularity_estimate', 'danceability', 'energy', 'instrumentalness']].groupby('track_id', as_index=False).mean()
df_track_details_qualitative_variables = df_track_details[['track_id', 'release_year', 'mode_numeric']].groupby('track_id', as_index=False).median()
df_track_details = pd.merge(df_track_details_quantitative_variables, df_track_details_qualitative_variables, on='track_id')
```

```python
# Transformando o mode novamente para seus valores originais
df_track_details['mode_numeric'] = df_track_details['mode_numeric'].map({0.0: 'Menor', 1.0:'Maior'})
```

```python
# Truncando parte decimal de vari√°veis booleanas
df_track_details['release_year'] = df_track_details['release_year'].astype(int)
```

```python
# Dict com a tradu√ß√£o dos headers do dataset
translated_header_names = {
    'track_id': 'ID da m√∫sica',
    'duration': 'Dura√ß√£o (em segundos)',
    'us_popularity_estimate': 'Estimativa de popularidade nos EUA',
    'danceability': 'Dan√ßabilidade', 
    'energy': 'Energia', 
    'instrumentalness':'Instrumentalidade', 
    'release_year': 'Ano de Lan√ßamento', 
    'mode_numeric': 'Escala musical'
}
```

```python
df_track_details.rename(columns=translated_header_names, inplace=True)
```

```python
df_track_details
```

# Manipulando Users details

```python
# Lendo o dataset dos detalhes de usu√°rio
df_users_details = pd.read_csv('datasets/source/users_details.csv')
```

```python
df_users_details
```

```python
## Conferindo se h√° colunas com valor nulo e o tipo dos dados de cada coluna
df_users_details.info()
```

```python
# Separando as vari√°veis que usaremos (valores baseados na moda)
df_users_details = df_users_details[['session_id', 'track_id_clean', 'skip_1', 'not_skipped', 'hour_of_day']].groupby('track_id_clean', as_index=False).median()
```

```python
# Truncando parte decimal de vari√°veis booleanas
df_users_details['skip_1'] = df_users_details['skip_1'].astype(bool)
df_users_details['not_skipped'] = df_users_details['not_skipped'].astype(bool)
df_users_details['hour_of_day'] = df_users_details['hour_of_day'].astype(int)
```

```python
# Transformando valores num√©ricos em booleanos
df_users_details['skip_1'] = df_users_details['skip_1'].map({False: 'Falso', True:'Verdadeiro'})
df_users_details['not_skipped'] = df_users_details['not_skipped'].map({False: 'Falso', True:'Verdadeiro'})
```

```python
# Dict com a tradu√ß√£o dos headers do dataset
translated_header_names = {
    'track_id_clean': 'ID da m√∫sica',
    'skip_1': 'M√∫sica foi pulada no in√≠cio',
    'not_skipped': 'M√∫sica n√£o foi pulada',
    'hour_of_day': 'Escutada em qual hora do dia',
}
```

```python
df_users_details.rename(columns=translated_header_names, inplace=True)
```

```python
df_users_details
```

# Unindo Datasets

```python
# Une os datasets com base no campo 'ID da m√∫sica'
df = pd.merge(df_track_details, df_users_details, on='ID da m√∫sica')
```
